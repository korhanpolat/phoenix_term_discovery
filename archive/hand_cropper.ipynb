```python
import numpy as np
import os
import cv2
```

```python
def get_hand(x,y,frame):
    x = int(x)
    y = int(y)
    
    if x < 46:
        x = 46
    if x > 210 - 46:
         x = 210 - 46
    if y < 66:
        y = 66
    if y > 260-66:
        y = 260 - 66
    f = frame[y-66:y+66,x-46:x+46,:]
    f = cv2.resize(f,(92,132))
    return f
```

```python
npy_root = '/home/korhan/Desktop/tez/dataset/features/alphapose/train'
imgs_root = '/home/korhan/Desktop/tez/dataset/features/fullFrame-210x260px/train'

hand = 'left'
imgs_target = '/home/korhan/Desktop/tez/dataset/features/hand_crop/' + hand + '/train'


files_list = os.listdir(npy_root)
len(files_list)
```

```python
alpha_wrist = {'left': 9, 'right': 10}


for i in range(len(files_list)):
    print('{} / {}'.format(i,len(files_list)))
    feats = np.load(os.path.join(npy_root,files_list[i]))

#     n_frame = feats.shape[0]

    imgs_list = sorted(os.listdir(os.path.join(imgs_root, files_list[i][:-4])))

    try: os.mkdir(os.path.join(imgs_target, files_list[i][:-4]))
    except: print('dir exists')

    x,y = 0,0

    for k,img_name in enumerate(imgs_list):
        img = cv2.imread(os.path.join(imgs_root, files_list[i][:-4], img_name))
    #     conf threshold konulabilir
        x,y = feats[k,alpha_wrist[hand],:2]
        crop = get_hand(x,y,img)
        cv2.imwrite(os.path.join(imgs_target, files_list[i][:-4], img_name), crop)
```
